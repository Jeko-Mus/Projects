```python
import pandas as pd
import requests
from bs4 import BeautifulSoup   
```

```python
import requests
page = requests.get("https://www.imdb.com/search/title/?title_type=feature&num_votes=25000,&genres=crime&sort=user_rating,desc&ref_=adv_prv")
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
Title = []
Description = []
Release = []
Director = []
Rating = []
Duration = []
Genre = []
Actors = []
# Actors_updated
Filming = []

#look for titles and release years
content_1 = soup.find_all('h3', class_="lister-item-header")
for i in content_1:
    name = i.find('a').text
    Title.append(name)
    year = i.find('span', class_="lister-item-year text-muted unbold").text.strip('()I (')
    Release.append(year)


# look for duration, genre, description, rating
content_2 = soup.find_all('div', class_ ="lister-item-content")

for j in content_2:
    descr = j.find_all('p', class_="text-muted")[1].text.strip('\n')
    Description.append(descr)
    rate = j.find('strong').text
    Rating.append(rate)
    dura = j.find('span', class_="runtime").text.strip(' min')
    Duration.append(dura)
    genr = j.find('span', class_ ="genre").text.strip('\n            ')
    Genre.append(genr)

# look for directors and actors
content_3 = soup.find_all('p', class_='')

for k in content_3:
    direc = k.find_all('a')[0].text
    Director.append(direc)
    Actors.append(k.find_all('a')[1].text)
    Actors.append(k.find_all('a')[2].text)
    Actors.append(k.find_all('a')[3].text)
    Actors.append(k.find_all('a')[4].text)
    
#split list of 200 into lists of 4 for actors

Actors_updated = list()
list_size = 4
for i in range(0, len(Actors), 4):
    Actors_updated.append(Actors[i:i+4])

```
